```python
from datetime import datetime
df["Used Date"] = pd.to_datetime(df["Used Date"])
df["Year"] = df["Used Date"].dt.year
df["Month"] = df["Used Date"].dt.month
df["Day"] = df["Used Date"].dt.day
df["Weekday"] = df["Used Date"].dt.day_name()
```

```python
df["Arrival_Hour"] = df["Scheduled Arrival"].dt.hour
df["Arrival_Minute"] = df["Scheduled Arrival"].dt.minute
```

```python
numerical_cols = [
    "Distance",
    "Passenger Load Factor",
    "Airline Rating",
    "Airport Rating",
    "Market Share",
    "OTP Index",
    "weather__hourly__windspeedKmph",
    "weather__hourly__precipMM",
    "weather__hourly__humidity",
    "weather__hourly__visibility",
    "weather__hourly__pressure",
    "weather__hourly__cloudcover"
]

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    print(f"{col}: {outliers} outliers")
```

```python
#One-Hot Encoding
categorical_cols = [
    "From",
    "To",
    "Airline",
    "weather__hourly__weatherDesc__value",
    "Category",
    "Weekday"
]

df = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True,
    dtype=int
)
```

```python
print(df.shape)
df.head()
```

```python
X_dep = df.drop(columns=["Departure Delay", "Arrival Delay"])
y_dep = df["Departure Delay"]
```

```python
X_arr = df.drop(columns=["Departure Delay", "Arrival Delay"])
y_arr = df["Arrival Delay"]

```

```python
from sklearn.model_selection import train_test_split

X_train_dep, X_test_dep, y_train_dep, y_test_dep = train_test_split(
    X_dep,
    y_dep,
    test_size=0.2,
    random_state=42
)
```

```python
X_train_arr, X_test_arr, y_train_arr, y_test_arr = train_test_split(
    X_arr,
    y_arr,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr_dep = LinearRegression()
lr_dep.fit(X_train_dep, y_train_dep)

pred_lr_dep = lr_dep.predict(X_test_dep)
```

```python
lr_arr = LinearRegression()
lr_arr.fit(X_train_arr, y_train_arr)

pred_lr_arr = lr_arr.predict(X_test_arr)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


print("Departure Delay Model")
print("MAE :", mean_absolute_error(y_test_dep, pred_lr_dep))
print("RMSE:", np.sqrt(mean_squared_error(y_test_dep, pred_lr_dep)))
print("R² :", r2_score(y_test_dep, pred_lr_dep))
```

```python
print("Arrival Delay Model")
print("MAE :", mean_absolute_error(y_test_arr, pred_lr_arr))
print("RMSE:", np.sqrt(mean_squared_error(y_test_arr, pred_lr_arr)))
print("R² :", r2_score(y_test_arr, pred_lr_arr))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_dep = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_dep.fit(X_train_dep, y_train_dep)
pred_rf_dep = rf_dep.predict(X_test_dep)
```

```python
print("Random Forest - Departure Delay")
print("MAE :", mean_absolute_error(y_test_dep, pred_rf_dep))
print("RMSE:", np.sqrt(mean_squared_error(y_test_dep, pred_rf_dep)))
print("R² :", r2_score(y_test_dep, pred_rf_dep))
```

```python
rf_arr = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_arr.fit(X_train_arr, y_train_arr)

pred_rf_arr = rf_arr.predict(X_test_arr)

print("Random Forest - Arrival Delay")
print("MAE :", mean_absolute_error(y_test_arr, pred_rf_arr))
print("RMSE:", np.sqrt(mean_squared_error(y_test_arr, pred_rf_arr)))
print("R² :", r2_score(y_test_arr, pred_rf_arr))
```
